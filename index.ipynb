## some checks that should produce gray images (= null error)

null translation gives the same image

```python
plt.imshow(image - im.translate(image, [0., 0.]), **opts)
```

the lg.translate function is invertible

```python
plt.imshow(image - im.translate(im.translate(image, [1.64, -2.233]), [-1.64, 2.233]), **opts)
#plt.colorbar()
```

also true for bigger translations  (we separate integer part from fractional part in the translation)

```python
plt.imshow(image - im.translate(im.translate(image, [182.64, -286.233]), [-182.64, 286.233], preshift=True), **opts)
```

but not always true when we don't separate

```python
plt.imshow(image - im.translate(im.translate(image, [182.64, -286.233]), [-182.64, 286.233], preshift=False), **opts)
```

the lg.translate function is periodic

```python
plt.imshow(image - im.translate(image, [-45*N_X, 5*N_Y]), **opts)
```

lg.translate function is associative

```python
plt.imshow(im.translate(image, [1., .0]) - im.translate(im.translate(image, [.5, .0]), [.5, .0]), **opts)
```

```python
plt.imshow(im.translate(image, [2., .0]) - im.translate(im.translate(image, [1.5, -1.0]), [.5, 1.0]), **opts)
```

## Whitening

Testing the whitening strategy.

Generates 2 figures:

* one which compares Olshausen's formulation and the actual statistical measure
* one which shows the actual effect of whitening


```python
debug = True
debug = False

ext = '.pdf' # type of output figure

url_database = 'Yelmo'

if not(debug):
    #####################
    figures_corr = True
    figures_whitening = True
    figures_atick = True
    #####################
#        image = pylab.imread('database/11-02-23_bednar_20100602_13cx632_d900_shrew_juvi/00000180.jpg')[:,:,0]
    image = plt.imread('database/lena512.png')[:,:,0]
    n_learning = 200
    figure_height = 6.  #inches
    results_folder  = 'figures/'
    recompute = False
else:
    print ' /!\ DEBUGGING /!\ '
    ##################### True False
    figures_corr = True
    figures_whitening = False
    figures_atick = True
    #####################
    image = plt.imread('database/yelmo256.png')[:,:,0]
    n_learning = 10
    figure_height = 6.  #inches
    results_folder = 'figures/'
    recompute = True

pe.N_X, pe.N_Y = image.shape

im = Image(pe)
pe.datapath = '../AssoField/database/'

import os
if not(os.path.isdir(results_folder)):  os.mkdir(results_folder)

K = im.whitening_filt(size = image.shape,
                   url_database = url_database,
                   n_learning = n_learning*5,
                   recompute = recompute,
                   )
```

```python
# TODO: know what N means : figure 3 ?

def fig_corr(K, n_learning, url_database):
    global im, ext, figure_height, results_folder
    print 'Figure Corr'
    bord = .03
    
    middle = np.ceil(K.shape[0]/2)
    K_axis= np.arange(-middle,middle)
    vec = .5 * (K[middle,:] + K[middle+1,:])
    hor = .5 * (K[:,middle] + K[:,middle+1])

    K_ols = im.olshausen_whitening_filt()
    vec_ols = .5 * (K_ols[middle,:] + K_ols[middle+1,:])

    #axes(rect, axisbg='w') where rect=[left, bottom, width, height] in
    #normalized (0,1) units.  axisbg is the background color for the
    #axis, default white

    fig = plt.figure(figsize=(3*figure_height,figure_height))
    a = plt.axes([bord, bord, 1/3.-bord,  1-bord])
    a.semilogy(K_axis,1/vec,c='b')
    a.semilogy(K_axis,1/hor,c='g')
    a.semilogy(K_axis,1/vec_ols,c='k', ls='--',label='ols')
    a.semilogy([],c='k', ls='-',label='stat')
    a.legend(loc="upper right")

    a = plt.axes([1/3.+bord, bord, 1/3.-2*bord , 1. -2*bord])
    a.matshow(np.log(K))
    a.axis('off')

    a = plt.axes([2/3.+bord, bord, 1/3.-bord, 1-bord])
    a.semilogy(K_axis,vec,c='b')
    a.semilogy(K_axis,hor,c='g')
    a.semilogy(K_axis,vec_ols,c='k', ls='--')
    a.semilogy([],c='b',label='vert')
    a.semilogy([],c='g',label='hor')
    a.legend(loc="upper right")
    #fig.savefig(results_folder + '/'  + 'whitening_corr' + ext)
    #plt.close(fig)
    plt.show()
    
fig_corr(K,n_learning,url_database)
```

```python
import pyprind
def fig_whitening(image,K,n_learning,url_database):
    global pe, ext, figure_height, results_folder

    try:
        xcorr = np.load(os.path.join(results_folder,'mat-xcorr'))#results['xcorr']
        xcorr_white = np.load(os.path.join(results_folder,'mat-xcorr-white'))#results['xcorr_white']
    except:
        print ' Testing that whitening effectively works... '
        size_patch = (128,128)
        pe.N_X =128
        im_patch = Image(pe)
        xcorr, xcorr_white = np.zeros(size_patch), np.zeros(size_patch)
        bar = pyprind.ProgBar(n_learning)
        for i_learning in range(n_learning):

            image_patch, filename, croparea = im_patch.patch(url_database, verbose=False)
            xcorr += im_patch.coco(image_patch, image_patch)/n_learning
            image_patch = im_patch.whitening(image_patch)
            xcorr_white += im_patch.coco(image_patch, image_patch)/n_learning
            bar.update()
        print(bar)
        np.save(os.path.join(results_folder,'mat-xcorr'), xcorr)
        np.save(os.path.join(results_folder,'mat-xcorr-white'), xcorr_white)
    #results.close()


    middle = np.ceil(K.shape[0]/2)
#    xcorr_middle = .5 * (K[middle,:] + K[middle+1,:])


    print 'Figure whitening'
    fig = plt.figure(figsize=(1.618*figure_height,figure_height))
    bord = .03
    # top left
    a = plt.axes([0,  1/2., 1/3., 1/2.])
    a.matshow(image, cmap=pylab.gray())
    a.axis('off')
    # top right
    a = plt.axes([2/3., 1/2., 1/3., 1/2.])
    white = im.retina(image)
    a.matshow(white, cmap=pylab.gray())
    a.axis('off')
    # middle
    a = plt.axes([1/3.+bord, .25, 1/3.-2*bord , 1/2.])
    impulse = np.zeros(image.shape)#(25,25))
    impulse[24,24] = 1
    K = im.retina(impulse)[:49,:49]
    middle = np.ceil(K.shape[0]/2)
    K_middle = .5 * (K[middle,:] + K[middle+1,:])
    a.plot([24,24],[np.min(K_middle), np.max(K_middle)], c='k')
    a.plot([0, 49],[0,0], c='k')
    a.plot(K_middle, c='b')
    a.axis('off')
    # bottom left
    a = plt.axes([0,  0, 1/3., 1/2.])
    a.matshow(xcorr, cmap=pylab.jet())#
    a.axis('off')
    #bottom right
    a = pylab.axes([2/3., 0, 1/3., 1/2.])
    a.matshow(xcorr_white, cmap=pylab.jet())
    a.axis('off')
    # gid save the figure
    #fig.savefig(results_folder + '/'  + 'whitening' + ext)
    #plt.close(fig)
    plt.show()

fig_whitening(image,K,n_learning,url_database)
```

```python
def fig_atick(K):
    """ This is just to remember that we use a simpler fitering technique.

    In the Attick LGN, the gain is changed according to an estimation of the SNR.

    """

    global ext,figure_height,results_folder
    print 'Figure Atick'

    contrasts = 1. / 2**np.arange(6)
    freqs = np.linspace(0.,.49, 40)
    size  = K.shape
    x, y = np.mgrid[0:size[0],0:size[1]]

    response = np.zeros((len(freqs), len(contrasts)))
    for i_contrast, contrast in enumerate(contrasts):
        for i_freq, freq in enumerate(freqs):
            image = contrast * np.sin( 2* np.pi * x * freq)
            white = im.whitening(image)
            response[i_freq,i_contrast] = np.std(white)
    response.shape

    fig = pylab.figure(figsize=(1.618*figure_height,figure_height))
    a = pylab.subplot(111)
    a.loglog(freqs,response,c='b')
    pylab.xlabel('frequency (cycles / pixel)')
    pylab.axis('tight')
    #fig.savefig(results_folder + '/'  + 'whitening_atick' + ext)
    #pylab.close(fig)
    pylab.show()

fig_atick(K)
```
